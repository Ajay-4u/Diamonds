```python
diamond = diamond.rename(columns = {'carat' : 'carat', \
'cut' : 'cut', \
'color' : 'color', \
'clarity' : 'clarity', \
'depth' : 'depth(%)', \
'table' : 'table', \
'x' : 'length', \
'y' : 'width', \
'z' : 'depth(mm)', \
'price' : 'price'})
diamond.head()
```

```python
diamond = diamond.loc[:, ['carat', 'cut', 'color', 'clarity', 'depth(%)', 'table', 'length',
       'width', 'depth(mm)', 'price']]
diamond.head()
```

```python
# data conversion for color column
j = diamond[diamond['color']=='J']
j.loc[:, 'color'] = 1

i = diamond[diamond['color']=='I']
i.loc[:, 'color'] = 2

h = diamond[diamond['color']=='H']
h.loc[:, 'color'] = 3

g = diamond[diamond['color']=='G']
g.loc[:, 'color'] = 4

f = diamond[diamond['color']=='F']
f.loc[:, 'color'] = 5

e = diamond[diamond['color']=='E']
e.loc[:, 'color'] = 6

d = diamond[diamond['color']=='D']
d.loc[:, 'color'] = 7

diamond1= pd.concat([j, i, h, g, f, e, d])
```

```python
# data conversion for cut column
fair = diamond1[diamond1['cut']=='Fair']
fair.loc[:, 'cut'] = 1

good = diamond1[diamond1['cut']=='Good']
good.loc[:, 'cut'] = 2

vGood = diamond1[diamond1['cut']=='Very Good']
vGood.loc[:, 'cut'] = 3

prem = diamond1[diamond1['cut']=='Premium']
prem.loc[:, 'cut'] = 4

ideal = diamond1[diamond1['cut']=='Ideal']
ideal.loc[:, 'cut'] = 5

diamond2= pd.concat([fair, good, vGood, prem, ideal])
```

```python
# data conversion for clarity column
I1 = diamond2[diamond2['clarity']=='I1']
I1.loc[:, 'clarity'] = 1

SI2 = diamond2[diamond2['clarity']=='SI2']
SI2.loc[:, 'clarity'] = 2

SI1 = diamond2[diamond2['clarity']=='SI1']
SI1.loc[:, 'clarity'] = 3

VS2 = diamond2[diamond2['clarity']=='VS2']
VS2.loc[:, 'clarity'] = 4

VS1 = diamond2[diamond2['clarity']=='VS1']
VS1.loc[:, 'clarity'] = 5

VVS2 = diamond2[diamond2['clarity']=='VVS2']
VVS2.loc[:, 'clarity'] = 6

VVS1 = diamond2[diamond2['clarity']=='VVS1']
VVS1.loc[:, 'clarity'] = 7

IF = diamond2[diamond2['clarity']=='IF']
IF.loc[:, 'clarity'] = 8

diamond3= pd.concat([I1, SI2, SI1, VS2, VS1, VVS2, VVS1, IF])
```
